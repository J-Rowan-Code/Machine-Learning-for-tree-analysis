# Algorithms explained


## Algorithims for classification 

### Logistic Regression

This is used to estimate discrete values (usually binary values like 0/1) from a set of independent variables. It helps predict the probability of an event by fitting data to a logit function. It is also called logit regression.

In this way we can set our classes at discrete values, then test the variable against these classifications. This is predicted to be quite ineffective as classes increases, but may offer a semblance of correct clssification for smaller classification levels

These methods listed below are often used to help improve logistic regression models:

- include interaction terms
- eliminate features
- regularize techniques
- using a non-linear model

### Naive Bayes 

A Naive Bayes classifier is based of Bayesian statistical process to make classifications, but makes the assumption that the presence of a particular feature in a class is unrelated to the presence of any other feature, meaning a reduced requirement of computational power, and an overall faster processing time.

Even if these features are related to each other, a Naive Bayes classifier would consider all of these properties independently when calculating the probability of a particular outcome.

A Naive Bayesian model is therefore easy to build and useful for massive datasets - perfectly appropriate for our use case. It's simple and is known to outperform even highly sophisticated classification methods. Our initial prediction is it will perform reasonably well against the other classification algorithms applied in this assignment.


### KNN Regression

This can be used for both regression and classification, but in this case we shall use it to store all available cases and classifies any new cases by taking a majority vote of its k neighbors. The case will then be assigned to the class with which it has the most in common. A distance function performing this measurement. This means our datasets will need to be normalized and standardised in order to prevent feature bias. 

KNN may perform better for where there are more classifications, such as by species, but will be very computationally expensive, (especially compared to methods such as Naive bayes. Feature slection will be very important to ensure KNN regression is appropriate.

Things to consider before selecting KNN: 

- KNN is computationally expensive
- Variables should be normalized, or else higher range variables can bias the algorithm
- Data still needs to be pre-processed.



### Support Vector Machine

SVM is a method of classification in which you plot raw data as points in an n-dimensional space (where n is the number of features you have). In a way, we can compare it to the lines used iteratively by ada boost for example.

The value of each feature is then tied to a particular coordinate, making it easy to classify the data. Lines called classifiers can be used to split the data and plot them on a graph.

SVM is of particular use when comparing large datasets, and may be most approriate for classifying things such as species (in the same way that KNN may perform well)


### Decision Tree

In this method we use a decision tree known as a classification tree. In this structure, leaves represent class labels and branches represent conjunctions of features that lead to those class labels.


Advantages

Amongst other data mining methods, decision trees have various advantages:

- Simple to understand and interpret. People are able to understand decision tree models after a brief explanation. Trees can also be displayed graphically in a way that is easy for non-experts to interpret.
- Able to handle both numerical and categorical data.
- Requires little data preparation. Other techniques often require data normalization. Since trees can handle qualitative predictors, there is no need to create dummy variables.
- Possible to validate a model using statistical tests. That makes it possible to account for the reliability of the model.
- Non-statistical approach that makes no assumptions of the training data or prediction residuals; e.g., no distributional, independence, or constant variance assumptions
- Performs well with large datasets. Large amounts of data can be analyzed using standard computing resources in reasonable time.
- Robust against co-linearity, particularly boosting
- In built feature selection. Additional irrelevant feature will be less used so that they can be removed on subsequent runs. The hierarchy of attributes in a decision tree reflects the importance of attributes. It means that the features on top are the most informative.


Limitations

- Trees can be very non-robust. A small change in the training data can result in a large change in the tree and consequently the final predictions.
- The problem of learning an optimal decision tree is known to be NP-complete under several aspects of optimality and even for simple concepts.
- Decision-tree learners can create over-complex trees that do not generalize well from the training data. (This is known as overfitting. Mechanisms such as pruning are necessary to avoid this problem (with the exception of some algorithms such as the Conditional Inference approach, that does not require pruning).
- For data including categorical variables with different numbers of levels, information gain in decision trees is biased in favor of attributes with more levels.However, the issue of biased predictor selection is avoided by the Conditional Inference approach,a two-stage approach,or adaptive leave-one-out feature selection.


### Random Forest


In this method we will use a collective of decision trees called a Random Forest. To classify a new object based on its attributes, each tree is classified, and the tree “votes” for that class. The forest chooses the classification having the most votes (over all the trees in the forest).

Each tree is planted & grown as follows:

- If the number of cases in the training set is N, then a sample of N cases is taken at random. This sample will be the training set for growing the tree.
- If there are M input variables, a number m<<M is specified such that at each node, m variables are selected at random out of the M, and the best split on this m is used to split the node. The value of m is held constant during this process.
- Each tree is grown to the most substantial extent possible. __There is no pruning necessary__. 


## Algorithims /Methods for feature selection


### Stepwise Regression

This is used in feature selection, acting to remove features in a dataset that exhibit multicollinearity . This will helps us select predictor variables in our models with less multicollinearity, based off their relevance from f-test statistic hypothesis testing.

This will be of particular use in Method 3, where the majority of our efforts will be towards.


### Chi Square Selection


The chi-square test is a statistical test of independence to determine the dependency of two variables. It shares similarities with coefficient of determination, R². However, chi-square test is only applicable to categorical or nominal data while R² is only applicable to numeric data.

We can therefor use this for feature selection

We have a target variable (i.e., the class label) and some other features (feature variables) that describe each sample of the data. Now, we calculate chi-square statistics between every feature variable and the target variable and observe the existence of a relationship between the variables and the target. If the target variable is independent of the feature variable, we can discard that feature variable. If they are dependent, the feature variable is very important.


### Entropy

A value between 0 and 1, a way of calculating lowest disorder (extremes at both end) as opposed to high disorder (the maximal value in middle of graph). In this method we are selecting for values with the lowest disorder




### Information Gain

This is a way of using Entropy, quantifying the size of uncertainty, disorder or impurity , where we should see a decrease in entropy from top (root) to the bottom (leaf nodes)

Information gain will be the product of probabilities of the class with a log having base 2 of that class probability



### Gini Index

Gini Index, also known as Gini impurity, is a value between 0 and 1 that calculates the amount of probability of a specific feature that is classified incorrectly when selected randomly. If all the elements are linked with a single class then it can be called pure.

0 expresses a purity of classification (ie only one clas or all elements belong to a specified class)
1 indicates random distribution of elements accross classes
0.5 indicates equal distribution



### Gini Index vs Information Gain

- The Gini Index facilitates the bigger distributions so easy to implement whereas the Information Gain favors lesser distributions having small count with multiple specific values.
- The method of the Gini Index is used by CART algorithms, in contrast to it, Information Gain is used in ID3, C4.5 algorithms.
- Gini index operates on the categorical target variables in terms of “success” or “failure” and performs only binary split, in opposite to that Information Gain computes the difference between entropy before and after the split and indicates the impurity in classes of elements.

```python
def calc_entropy(p):
    if p!=0:
        return -p*np.log2(p)
    else:
        return 0

def calc_entropy_array(classes):
    # List the possible class values
    newClasses = []
    for aclass in classes:
        if newClasses.count(aclass)==0:
            newClasses.append(aclass)
            
    #histogram
    frequency = np.zeros(len(newClasses))
    
    for value in classes:
        frequency[(value-1)] = frequency[(value-1)] +1
        totalEntropy=0.0
    
        
    for f in frequency:
        totalEntropy += calc_entropy(f/len(classes))
        
        
    return totalEntropy

def calc_informationgain(data,classes, feature):
    partial_gain = 0
    nData = len(data)
    #list the value that the feature can take
    values = []
    for datapoint in data:
        if datapoint[feature] not in values:
            values.append(datapoint[feature])
    
    featureCounts = np.zeros(len(values))

    valueIndex = 0
    #find where those values appear in data[feature] and the corresponding class
    for value in values:
        print("Still loopin")
        dataIndex = 0
        newClasses = []
        #print("value ",value)
        for datapoint in data:
            if datapoint[feature]==value:
                featureCounts[valueIndex]+=1
                newClasses.append(classes[dataIndex])
            dataIndex += 1

        #Get the values in newClasses
        classValues = []
        entropy = np.zeros(len(values))  
        for aclass in newClasses:
            if classValues.count(aclass)==0:
                classValues.append(aclass)
                classCounts = np.zeros(len(classValues))
                classIndex=0
                for classValue in classValues:
                    for aclass in newClasses:
                        if aclass == classValue:
                            classCounts[classIndex]+=1
                    classIndex += 1

        for classIndex in range(len(classValues)):
             entropy[valueIndex] += calc_entropy(float(classCounts[classIndex])/sum(classCounts))

        
        #============================
        # YOUR CODE HERE
        #============================
        #partial_gain = partial_gain + (entropy[valueIndex] * (featureCounts[valueIndex]/nData))
        partial_gain += float(featureCounts[valueIndex])/nData * entropy[valueIndex]

        valueIndex += 1
    
    return partial_gain
    
def TotalGini(classes):
    # List the possible class values
    newClasses = []
    for aclass in classes:
        if newClasses.count(aclass)==0:
            newClasses.append(aclass)
    #histogram
    frequency = np.zeros(len(newClasses))
    for value in classes:
        frequency[(value-1)]=frequency[(value-1)]+1
    Gini=0.0
    
    for f in frequency:
        #============================
        # YOUR CODE HERE
        #============================
        Gini += (f/len(classes))**2
        
    Gini=1-Gini
    
    return Gini

def calc_Giniinformationgain(data,classes, feature):
    Gf = 0
    nData = len(data)
    #list the value that the feature can take
    values = []
    for datapoint in data:
        if datapoint[feature] not in values:
            values.append(datapoint[feature])
    featureCounts = np.zeros(len(values))

    valueIndex = 0
    #find where those values appear in data[feature] and the corresponding class
    for value in values:
        print("Loop: ", value, " still truckin dawg")
        dataIndex = 0
        newClasses = []
        #print("value ",value)
        for datapoint in data:
            if datapoint[feature]==value:
                featureCounts[valueIndex]+=1
                newClasses.append(classes[dataIndex])
            dataIndex += 1
     
        #Get the values in newClasses
        classValues = []
        gini = np.zeros(len(values))  
        for aclass in newClasses:
            if classValues.count(aclass)==0:
                classValues.append(aclass)
                classCounts = np.zeros(len(classValues))
                classIndex=0
                for classValue in classValues:
                    for aclass in newClasses:
                        if aclass == classValue:
                            classCounts[classIndex]+=1
                    classIndex += 1
  
        for classIndex in range(len(classValues)):
            gini[valueIndex] += (classCounts[classIndex]/sum(classCounts))**2
      
        Gf += float(featureCounts[valueIndex])/nData * gini[valueIndex]
        valueIndex += 1
    return 1-Gf
```

```python
#dfName = Pandas dataframe that you wish to prune
#dataColumns = The columns up to which there is data in (will use all columns up to and including specified column number)
#-- Will not work if there are arbitrary string columns or uncleaned columns
#Class column = the index number of the column that contains the class labels
#Wanted Columns = The amount of columns you wish your returned dataframe to contain (will select columns with the best information scores)
def giniPruningFunction(dfName, dataColumns, classColumn, wantedColumns = 5):
    bestGiniList = []
    colGiniList = []
    tempArray = dfName.values
    targets = tempArray[:,classColumn]
    #print(targets)
    data = tempArray[:,:dataColumns]
    totalgini = TotalGini(targets)
    for x in range(len(data[0])):
        if x == classColumn:
            continue
        lowestGin = 1
        #print(x)
        ginigain = calc_Giniinformationgain(data,targets,x)
        ginigain = totalgini - ginigain
        if len(bestGiniList) < wantedColumns:
            bestGiniList.append(ginigain)
            colGiniList.append(x)
        else:
            for y in range(len(bestGiniList)):
                if lowestGin > bestGiniList[y]:
                    lowestGin = bestGiniList[y]
                    replaceCol = y
                    #print("low: ", lowestGin)
            if ginigain > lowestGin:
                bestGiniList[replaceCol] = ginigain
                colGiniList[replaceCol] = x
                #if ginigain > lowestGin:
                #    bestGiniList[y] = ginigain
                #    colGiniList[y] = x
    #print(bestGiniList)
    #giniDict = dict(zip(colGiniList, bestGiniList))
    colGiniList.append(classColumn)
    colGiniList.sort()
    returnDF = dfName[colGiniList]
    
    return returnDF
```

```python
#dfName = The variable name of the pandas dataframe you wish to prune
#Class column = the index number of the column that contains the class labels
#Wanted Columns = The amount of columns you wish your returned dataframe to contain (will select columns with the best information scores)
def entropyPruningFunction(dfName, classColumn, wantedColumns = 5):
    bestEntropy = []
    colList = []
    targets = dfName.loc[:,classColumn].values
    data = dfName.values
    total_entropy = calc_entropy_array(targets)
    for x in range(len((data[0]))):
        if x == classColumn:
            break
        lowestEntropy = 1
        partial_gain = calc_informationgain(data, targets, x)
        partial_gain = total_entropy - partial_gain
        if len(bestEntropy) < wantedColumns:
            bestEntropy.append(partial_gain)
            colList.append(x)
        else:
            for y in range(len(bestEntropy)):
                if lowestEntropy > bestEntropy[y]:
                    lowestEntropy = bestEntropy[y]
                    replaceCol = y
            if partial_gain > lowestEntropy:
                bestEntropy[replaceCol] = partial_gain
                colList[replaceCol] = x
    colList.append(classColumn)
    colList.sort()
    returnDF = dfName[colList]
    return returnDF
```
